```julia
using JuMP
using Ipopt
using Plots
using LinearAlgebra

# Problem parameters
T = 60  # Time horizon (seconds)
dt = 1/3600  # Time step in hours (1 second converted to hours)

# Initial conditions
xA_init = [0.0, 0.0]  # Alice's initial position (miles)
vA_init = [0.0, 20.0]  # Alice's initial velocity (mph)
xB_init = [0.5, 0.0]  # Bob's initial position (miles)
vB_init = [30.0, 0.0]  # Bob's initial velocity (mph)

function solve_rendezvous_with_velocity_matching()
    model = Model(Ipopt.Optimizer)
    set_silent(model)

    # Variables
    @variables(model, begin
        xA[1:T+1, 1:2]  # Alice's position over time
        vA[1:T+1, 1:2]  # Alice's velocity over time
        uA[1:T, 1:2]    # Alice's thrust inputs
        xB[1:T+1, 1:2]  # Bob's position over time
        vB[1:T+1, 1:2]  # Bob's velocity over time
        uB[1:T, 1:2]    # Bob's thrust inputs
    end)

    # Initial conditions constraints
    @constraints(model, begin
        [i=1:2], xA[1,i] == xA_init[i]
        [i=1:2], vA[1,i] == vA_init[i]
        [i=1:2], xB[1,i] == xB_init[i]
        [i=1:2], vB[1,i] == vB_init[i]
    end)

    # Dynamics constraints
    @constraints(model, begin
        # Alice's dynamics
        [t=1:T, i=1:2], xA[t+1,i] == xA[t,i] + dt * vA[t,i]
        [t=1:T, i=1:2], vA[t+1,i] == vA[t,i] + uA[t,i]
        
        # Bob's dynamics
        [t=1:T, i=1:2], xB[t+1,i] == xB[t,i] + dt * vB[t,i]
        [t=1:T, i=1:2], vB[t+1,i] == vB[t,i] + uB[t,i]
        
        # Rendezvous position constraint
        [i=1:2], xA[T+1,i] == xB[T+1,i]
        
        # Velocity matching constraint at rendezvous
        [i=1:2], vA[T+1,i] == vB[T+1,i]
    end)

    # Objective: Minimize total energy
    @objective(model, Min, 
        sum(uA[t,1]^2 + uA[t,2]^2 for t in 1:T) + 
        sum(uB[t,1]^2 + uB[t,2]^2 for t in 1:T)
    )

    optimize!(model)
    
    return value.(xA), value.(vA), value.(uA),
           value.(xB), value.(vB), value.(uB)
end

# Solve the problem
xA, vA, uA, xB, vB, uB = solve_rendezvous_with_velocity_matching()

# Calculate energies
energy_A = sum(uA[t,1]^2 + uA[t,2]^2 for t in 1:T)
energy_B = sum(uB[t,1]^2 + uB[t,2]^2 for t in 1:T)

# Plot results
p = plot(size=(1000,800))

# Position trajectories
plot!(p, xA[:,1], xA[:,2], label="Alice", linewidth=2)
plot!(p, xB[:,1], xB[:,2], label="Bob", linewidth=2)
scatter!(p, [xA[1,1]], [xA[1,2]], label="Alice Start", markersize=8)
scatter!(p, [xB[1,1]], [xB[1,2]], label="Bob Start", markersize=8)
scatter!(p, [xA[end,1]], [xA[end,2]], label="Rendezvous Point", markersize=10)

# Add velocity vectors at rendezvous
quiver!(p, [xA[end,1]], [xA[end,2]], 
        quiver=([vA[end,1]/50, vA[end,2]/50]), 
        label="Final Velocity")

xlabel!("X Position (miles)")
ylabel!("Y Position (miles)")
title!("Hovercraft Trajectories with Velocity Matching")

# Print results
println("\nResults with Velocity Matching:")
println("Rendezvous point: (", round(xA[end,1], digits=3), ", ", 
        round(xA[end,2], digits=3), ") miles")
println("Final velocity: (", round(vA[end,1], digits=3), ", ", 
        round(vA[end,2], digits=3), ") mph")
println("Alice's energy: ", round(energy_A, digits=3))
println("Bob's energy: ", round(energy_B, digits=3))

display(p)
```
